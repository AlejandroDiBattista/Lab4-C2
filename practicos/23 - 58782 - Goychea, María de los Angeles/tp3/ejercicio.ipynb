## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines()

    datos = []
    for linea in lineas:
        registro = {
            'fecha': linea[0:10].strip(),
            'producto': linea[10:40].strip(),
            'precio': float(linea[40:50].strip()),
            'cantidad': int(linea[50:55].strip())
        }
        datos.append(registro)
    
    return datos
datos = cargar_datos()

for registro in datos:
    print(registro)
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    total_importe = 0
    total_cantidad = 0
    
    for venta in datos:
        importe_venta = venta['precio'] * venta['cantidad']
        
        total_importe += importe_venta
        total_cantidad += venta['cantidad']
    return total_importe, total_cantidad

importe, cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
    ventas_por_producto = {}
    
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']

        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
            ventas_por_producto[producto] = cantidad
    return ventas_por_producto

def listar_ventas(ventas):
    print("Unidades vendidas por producto:")
    
    for producto, cantidad in ventas.items():
        print(f"{producto}: {cantidad} unidades")

ventas = unidades_vendidas(datos)
listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio(datos):
    suma_precios = {}
    total_cantidad = {}
    
    for venta in datos:
        producto = venta["producto"]
        precio = venta["precio"]
        cantidad = venta["cantidad"]
        
        if producto in suma_precios:
            suma_precios[producto] += precio * cantidad
            total_cantidad[producto] += cantidad
        else:
            suma_precios[producto] = precio * cantidad
            total_cantidad[producto] = cantidad
    
    promedios = {producto: suma_precios[producto] / total_cantidad[producto] for producto in suma_precios}
    return promedios

def listar_precios(promedios):
    for producto, precio in promedios.items():
        print(f"Producto: {producto}, Precio promedio: ${precio:.2f}")

precios = precio_promedio(datos)
listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    ventas = unidades_vendidas(datos)
    ranking = sorted(ventas.items(), key=lambda item: item[1], reverse=True)
    return ranking[:top]

def listar_ranking(ranking):
    for producto, cantidad in ranking:
        print(f"Producto: {producto}, Unidades vendidas: {cantidad}")

ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from collections import defaultdict
from datetime import datetime

def ventas_por_mes(datos):
    ventas_mensuales = defaultdict(lambda: defaultdict(int))
    
    for venta in datos:
        fecha = datetime.strptime(venta["fecha"], "%Y-%m-%d")
        mes = fecha.strftime("%Y-%m")
        producto = venta["producto"]
        cantidad = venta["cantidad"]
        ventas_mensuales[mes][producto] += cantidad
        
    return ventas_mensuales

def listar_ventas_mensuales(ventas):
    for mes, productos in ventas.items():
        print(f"Mes: {mes}")
        for producto, cantidad in productos.items():
            print(f"  Producto: {producto}, Unidades vendidas: {cantidad}")

ventas_mensuales = ventas_por_mes(datos)
listar_ventas_mensuales(ventas_mensuales)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):
    resumen = {}
    
    for registro in datos:
        producto = registro["producto"]
        precio = registro["precio"]
        cantidad = registro["cantidad"]
        
        if producto not in resumen:
            resumen[producto] = {
                "total_precio": 0,
                "total_cantidad": 0,
                "total_importe": 0,
                "transacciones": 0
            }
        resumen[producto]["total_precio"] += precio
        resumen[producto]["total_cantidad"] += cantidad
        resumen[producto]["total_importe"] += precio * cantidad
        resumen[producto]["transacciones"] += 1
    
    for producto, valores in resumen.items():
        valores["precio_promedio"] = valores["total_precio"] / valores["transacciones"]
    
    return resumen

def informe_ventas(resumen):
    productos_ordenados = sorted(resumen.keys())
    print(f"{'Producto':<30} {'Precio Promedio':<15} {'Unidades Vendidas':<18} {'Importe Total':<15}")
    print("="*80)
    
    for producto in productos_ordenados:
        valores = resumen[producto]
        precio_promedio = valores["precio_promedio"]
        cantidad_total = valores["total_cantidad"]
        importe_total = valores["total_importe"]
        print(f"{producto:<30} {precio_promedio:<15.2f} {cantidad_total:<18} {importe_total:<15.2f}")

datos = cargar_datos()  
resumen = resumen_ventas(datos)
informe_ventas(resumen)
````
