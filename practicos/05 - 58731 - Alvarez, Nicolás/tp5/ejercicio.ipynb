## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo,nombre,precio,tipo,cantidad):        
        self._codigo = self.validarcodigo(codigo);
        self._nombre = self.validarnombre(nombre);
        self._precio = self.validarprecio(precio)
        self._tipo = self.validartipo(tipo);
        self._cantidad = self.validarcantidad(cantidad);
    
    def validarprecio(self,precio):
         if int(precio)>=10 and int(precio)<=10000:
             return precio;
        
    def validartipo(self,tipo):
        longitud = len(str(tipo));
        if longitud >=0 and longitud<=20:
            return tipo;

    def valorTotal(self):
        return self.cantidad * self.precio;

    def validarcantidad(self,cantidad):
         if cantidad >=0 and cantidad <=100:
             return cantidad;

    def validarcodigo(self, codigo):
        longitud = len(str(codigo)) 
        if longitud == 4:
            return codigo;

    def validarnombre(self, nombre):
        longitud = len(str(nombre))  
        if longitud>=1 and longitud <=100:
            return nombre;
    
    @property
    def nombre(self):
        return self._nombre;

    @property
    def codigo(self):
        return self._codigo;

    @property
    def tipo(self):
        return self._tipo;

    @property
    def precio(self):
        return self._precio;
    
    @property
    def cantidad(self):
        return self._cantidad;

    @precio.setter
    def precio(self, value):
        if int(value)>=10 and int(value)<=10000:
         self._precio = value

    @cantidad.setter
    def cantidad(self, value):
        if value>=0 and value <=100:
         self._cantidad = value

    def _str_(self):
       if self.valorTotal is not None and self.codigo is not None and self.nombre is not None and self.precio is not None and self.cantidad is not None and self.tipo is not None:
            return f"Codigo valido: {self.codigo} y Nombre vlido: {self.nombre} y precio: {self.precio} y cantidad {self.cantidad} y tipo {self.tipo} y total {self.calcularvalortotal()}"
       else:
            return "Producto invalido"
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion="", codigos=[], tipos=[]):
        self.codigos = codigos;
        self.descripcion = descripcion;
        self.tipos = tipos;

    def validardescripcion(self, descripcion):
        longitud = len(str(descripcion));
        if longitud>=1 and longitud<= 100:
            return descripcion;

    def obtenerdescuento(self):
        return 0;

    def esAplicable(self,producto):
        # Verifica si el producto es aplicable según el código o el tipo
        return producto.codigo in self.codigos or producto.tipo in self.tipos;
             
    def aplicar(self, producto, cantidad):
        return producto.precio * cantidad

class OfertaDescuento(Oferta):
    def __init__(self,descuento, descripcion = "Descuento 10%" , *args, **kwargs):
            super().__init__(descripcion,*args, **kwargs);
            self.descuento = self.validardescuento(descuento)

    def validardescuento(self, descuento):
        if descuento>0 and descuento<= 100:
            return descuento

    def calcularDescuento(self,producto,cantidad):
        if not self.esAplicable(producto):
           return 0;
        total = producto.precio * cantidad;
        descuento_aplicado = total * (self.descuento / 100)
        return descuento_aplicado

    def __repr__(self):
        return f"Descuento de {self.descuento}%, Codigos: {self.codigos}";

class Oferta2x1(Oferta):
    def __init__(self, descripcion="Oferta 2x1", *args, **kwargs):
        super().__init__(descripcion, *args, **kwargs)

    def calcularDescuento(self,producto,tipos): 
       if not self.esAplicable(producto):
            return 0
        # Descuento basado en la oferta 2x1
       descuento_total = producto.precio * (tipos // 2)
       return descuento_total;
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv

class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []

    @staticmethod
    def leer(archivo):
        catalogo = Catalogo()
        with open(archivo, 'r') as archivo:
            lector = csv.reader(archivo)
            next(lector) 
            for fila in lector:
                codigo, nombre, precio, tipo, cantidad = fila
                producto = Producto(codigo, nombre, float(precio), tipo, int(cantidad))
                catalogo.agregar(producto)
        return catalogo

    def vender(self,producto,cantidad):
            for producto in self.productos:   
                if producto.codigo == producto.codigo and producto.cantidad >= cantidad:
                    producto.cantidad -= cantidad
                    break
            return self.productos;
    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(p.cantidad for p in self.productos)

    @property
    def valorTotal(self):
        return sum(p.precio * p.cantidad for p in self.productos)
    
    @property
    def precioPromedio (self):
            return self.valorTotal / self.cantidadProductos if self.cantidadProductos > 0 else 0

    def guardar(self, archivo):
        with open(archivo, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Codigo', 'Nombre', 'Precio', 'Tipo', 'Cantidad'])
            for producto in self.productos:
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])

    def agregar(self, *producto):
        for prod in producto:
            self.productos.append(prod)

    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto;

    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    def buscarOferta(self, producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto):
                return oferta

    def calcularDescuento(self, producto, cantidad):
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad);
        else:
            return 0;


    def informe(self):
        tipos = set(producto.tipo for producto in self.productos)
        informe = (
        f"Cantidad de productos: {self.cantidadProductos}\n"
        f"Cantidad de unidades: {self.cantidadUnidades}\n"
        f"Precio Promedio: {self.valorTotal / self.cantidadUnidades if self.cantidadUnidades > 0 else 0:.2f}\n"
        f"Valor total: {self.valorTotal}\n"
        f"Tipos de productos: {', '.join(tipos)}\n"
        f"Ofertas: {', '.join([o.descripcion for o in self.ofertas])}\n"
    )
        return informe
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente:
    def __init__(self, nombre, cuit):
        self._nombre = self.validarnombre(nombre)
        self._cuit = self.validarcuit(cuit)
    
    @property
    def nombre(self):
        return self._nombre
    
    @property
    def cuit(self):
        return self._cuit
    
    def validarnombre(self, nombre):
        longitud = len(str(nombre))
        if longitud >= 1 and longitud <= 100:
            return nombre
    
    def validarcuit(self, cuit):
        cuit = str(cuit)
        if len(cuit) == 13 and cuit[2] == '-' and cuit[11] == '-':
            partes = cuit.split('-')
            if len(partes) == 3 and all(part.isdigit() for part in partes):
                return cuit
    
    @nombre.setter
    def nombre(self, valor):
        if self.validarnombre(valor):
            self._nombre = valor
    
    @cuit.setter
    def cuit(self, valor):
        if self.validarcuit(valor):
            self._cuit = valor
```

```python
class Factura:
   ultimo_numero_factura = 100;

   def __init__(self, catalogo, cliente): 
        self.catalogo = catalogo;
        self.subtotal = 0.0;
        self.descuentos = 0.0;
        self.total = 0.0;
        self.cliente = cliente;
        self.productos = [];
        self.numero = Factura.nuevoNumero();
        self.cantidadProductos = 0;
        self.cantidadUnidades = 0
    
   @classmethod
   def ultimaFactura(cls, numero):
        cls.ultimo_numero_factura = numero;
    
   @classmethod
   def nuevoNumero(cls):
       cls.ultimo_numero_factura += 1
       return cls.ultimo_numero_factura
   
   def agregar(self, producto, cantidad):
        if self.catalogo.vender(producto, cantidad): 
                for i, (prod, cant) in enumerate(self.productos):
                    if prod.codigo == producto.codigo:
                        self.productos[i] = (prod, cant + cantidad)
                        self.subtotal += producto.precio * cantidad;
                        self.cantidadUnidades += cantidad;
                        self.aplicar_descuentos()
                        return
                self.productos.append((producto, cantidad));
                self.subtotal += producto.precio * cantidad;
                self.cantidadProductos += 1;
                self.cantidadUnidades += cantidad;
                self.aplicar_descuentos();

   def aplicar_descuentos(self):     
            self.descuentos = 0.0
            for producto, cantidad in self.productos:
               oferta = self.catalogo.buscarOferta(producto, cantidad)
            if oferta:
                    descuento = oferta.calcularDescuento(producto, cantidad)
                    self.descuentos += descuento
            self.total = self.subtotal - self.descuentos


   def imprimir(self):
    detalles_factura = f"Factura #{self.numero} - Cliente: {self.cliente.nombre}\n"
    for producto, cantidad in self.productos:
        detalles_factura += f"{cantidad} {producto.nombre} - {producto.precio * cantidad:.2f}\n"
        oferta = self.catalogo.buscarOferta(producto, cantidad)
        if oferta:
            if isinstance(oferta, Oferta2x1):
                detalles_factura += "Oferta 2x1 aplicada\n"
            if isinstance(oferta, OfertaDescuento):
                detalles_factura += "Descuento 10% aplicado\n"
    
    detalles_factura += f"Subtotal: {self.subtotal:.2f}\n"
    detalles_factura += f"Descuentos: {self.descuentos:.2f}\n"
    detalles_factura += f"TOTAL: {self.total:.2f}\n"
    
    return detalles_factura
```
