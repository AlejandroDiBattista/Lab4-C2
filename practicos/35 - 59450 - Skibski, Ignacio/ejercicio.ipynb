## Ejercicio 2

Escribe una función en Python que encuentre los valores de `a`, `b`, y `c` para que la función cuadrática `f(x) = aX^2 + bX + c` pase exactamente por los siguientes puntos:

| x  | y  |
|---:|---:|
|  0 |  0 |
|  1 |  8 |
|  2 | 12 |
|  3 | 12 |
|  5 |  0 |

### Requisitos:
- La función debe explorar posibles valores de `a`, `b`, y `c` utilizando un método de prueba y error.
- Debe devolver los valores que hagan que la diferencia entre la función `f(x)` y los valores medidos `y` sea exactamente cero para cada punto.

> **Pista**: Los valores de `a`, `b`, y `c` son números pequeños.

````python
#1. Tomar una lista de frases y convertir cada frase en una lista de palabras.
lista_frases = ["Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"]

def mostrar(lista_frases):
    lista_palabras = [frase.split() for frase in lista_frases]
    return lista_palabras

resultado = mostrar(lista_frases)
print(resultado)
````

````python
#2. Luego, debe crear una nueva lista que contenga la cantidad de palabras en cada frase.

lista_frases = ["Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"]

def mostrar(lista_frases):
    cantidad_palabra = []
    lista_cantidad_palabras = []

    for frase in lista_frases:
        cantidad_palabra = len(frase.split())
        lista_cantidad_palabras.append(cantidad_palabra)

    return lista_cantidad_palabras

resultado = mostrar(lista_frases)
print(resultado)
````

````python
#3. Tambien debe crear un lista que contenga la cantidad de caracteres en cada frase.

lista_frases = ["Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"]

def mostrar(lista_frases):
    cantidad_palabra = []
    lista_cantidad_palabras = []

    for frase in lista_frases:
        cantidad_palabra = len(frase)
        lista_cantidad_palabras.append(cantidad_palabra)

    return lista_cantidad_palabras

resultado = mostrar(lista_frases)
print(resultado)
````

````python
#4. Finalmente, debe imprimir cada frase original junto con la cantidad de palabras que contiene.

lista_frases = ["Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"]

def mostrar(lista_frases):
    lista_cantidad_palabras = []

    for frase in lista_frases:
        cantidad_palabras = len(frase.split())  
        lista_cantidad_palabras.append(cantidad_palabras)
        print(f'"{frase}" contiene {cantidad_palabras} palabras')  

    return lista_cantidad_palabras

resultado = mostrar(lista_frases)
````
