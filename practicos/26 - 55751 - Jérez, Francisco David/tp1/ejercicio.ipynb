## Ejercicio 1 

Escribe una función en Python que procese una lista de cadenas de texto y realice las siguientes acciones:

1. Tomar una lista de frases y convertir cada frase en una lista de palabras.
1. Luego, debe crear una nueva lista que contenga la cantidad de palabras en cada frase.
1. Tambien debe crear un lista que contenga la cantidad de caracteres en cada frase.
1. Finalmente, debe imprimir cada frase original junto con la cantidad de palabras que contiene.
Ejemplo de uso:
```python
    frases = [
        "Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"
    ]

    procesar_frases(frases)
```

ejemplo de salida:
```bash
    La frase: "Python es un..."
    tiene 6 palabras y 38 caracteres
```


````python
def imprimir_frases(frases):
    for frase in frases:
        palabras = frase.split()
        
        contar_palabra = len(palabras)
        contar_letras = len(frase)
        
        # Imprimimos el resultado
        print(f'La frase: "{frase}"')
        print(f'tiene {contar_palabra} palabras y {contar_letras} caracteres\n')

frases = [
    "me llamo Francisco",
    "me gusta este mundo de la programacion",
    "quiero aprender mas de python"
]

imprimir_frases(frases)
````

## Ejercicio 2

Escribe una función en Python que encuentre los valores de `a`, `b`, y `c` para que la función cuadrática `f(x) = aX^2 + bX + c` pase exactamente por los siguientes puntos:

| x  | y  |
|---:|---:|
|  0 |  0 |
|  1 |  8 |
|  2 | 12 |
|  3 | 12 |
|  5 |  0 |

### Requisitos:
- La función debe explorar posibles valores de `a`, `b`, y `c` utilizando un método de prueba y error.
- Debe devolver los valores que hagan que la diferencia entre la función `f(x)` y los valores medidos `y` sea exactamente cero para cada punto.

> **Pista**: Los valores de `a`, `b`, y `c` son números pequeños.

````python
def buscar_resultados():
    
    for a in range(-10, 11):
        for b in range(-10, 11):
            for c in range(-10, 11):
                if (a*0**2 + b*0 + c == 0 and
                    a*1**2 + b*1 + c == 8 and
                    a*2**2 + b*2 + c == 12 and
                    a*3**2 + b*3 + c == 12 and
                    a*5**2 + b*5 + c == 0):
                    return a, b, c

A, B, C = buscar_resultados()
print(f"a={A}; b={B}; c={C}")
print("espero que este bien profe!!! :)")
````
