## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_data():
    lineas = open("datos.dat", "r").readlines()
    data = []

    for linea in lineas:
        fecha = linea[:10].strip()
        producto = linea[10:40].strip()
        precio = int(linea[40:50].strip())
        cantidad = int(linea[50:].strip())

        data.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})
        
    print(data)
    return data

datos = cargar_data()

````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python




def calculo_totales(datos):
    lineas = open("datos.dat", "r").readlines()    

    importe = 0
    cantidad_total = 0

    for linea in lineas:
        precio = int(linea[40:50].strip())
        cantidad = int(linea[50:].strip())

        importe += precio * cantidad
        cantidad_total += cantidad

    return importe, cantidad_total

importe, cantidad_total = calculo_totales(datos)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad_total} unidades")



````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python

def unidad_vendidas(datos):
    ventas_por_producto = {}

    for linea in datos:
        producto = linea[10:40].strip()  
        cantidades = int(linea[50:55].strip())  

        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidades
        else:
            ventas_por_producto[producto] = cantidades
    return ventas_por_producto

def lista_ventas(ventas):
    for producto, cantidades in ventas.items():
        print(f'Producto: {producto}, Unidades vendidas: {cantidades}')

datos = open("datos.dat", "r").readlines()
ventas = unidad_vendidas(datos)
lista_ventas(ventas)

````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precios_promedios(datos):
    suma_precios = {}
    suma_cantidades = {}

    for linea in datos:
        producto = linea[10:40].strip()  
        precio = float(linea[40:50].strip())  
        cantidad = int(linea[50:55].strip())  

        if producto in suma_precios:
            suma_precios[producto] += precio * cantidad
            suma_cantidades[producto] += cantidad
        else:
            suma_precios[producto] = precio * cantidad
            suma_cantidades[producto] = cantidad

    precios_promedio = {}
    for producto in suma_precios:
        precios_promedio[producto] = suma_precios[producto] / suma_cantidades[producto]

    return precios_promedio

def listado_precios(precios):
    for producto, promedio in precios.items():
        print(f'Producto: {producto}, Precio promedio: {promedio:.2f}')

datos = open("datos.dat", "r").readlines()

precios = precios_promedios(datos)

listado_precios(precios)

````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def top3_productos(datos, top=3):
    ventas_de_producto = {}
    for linea in datos:
        producto = linea[10:40].strip()  
        cantidad = int(linea[50:55].strip())  

        if producto in ventas_de_producto:
            ventas_de_producto[producto] += cantidad
        else:
            ventas_de_producto[producto] = cantidad

    ranking = sorted(ventas_de_producto.items(), key=lambda x: x[1], reverse=True)

    return ranking[:top]

def listado_ranking(ranking):
    print("Ranking de productos más vendidos:")
    for i, (producto, cantidad) in enumerate(ranking, start=1):
        print(f"{i}. Producto: {producto}, Unidades vendidas: {cantidad}")

datos = open("datos.dat", "r").readlines()

ranking = top3_productos(datos)

listado_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python




def ventas_del_mes(datos):
    ventas_mensuales = {}
    for linea in datos:
        fecha = linea[:10].strip() 
        mes = fecha[:7]  
        producto = linea[10:40].strip()  
        cantidad = int(linea[50:55].strip())

        if mes not in ventas_mensuales:
            ventas_mensuales[mes] = {}
        if producto not in ventas_mensuales[mes]:
            ventas_mensuales[mes][producto] = 0
            
        ventas_mensuales[mes][producto] += cantidad
    return ventas_mensuales

def listar_ventas_mensuales(ventas):
    for mes, productos in ventas.items():
        print(f"Mes: {mes}")
        for producto, cantidad in productos.items():
            print(f"  Producto: {producto}, Unidades vendidas: {cantidad}")

datos = open("datos.dat", "r").readlines()
ventas = ventas_del_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python


def resumen_de_ventas(datos):
    ventas_resumen = {}
    for linea in datos:
        producto = linea[10:40].strip()  
        precio = float(linea[40:50].strip())  
        cantidad = int(linea[50:55].strip()) 
        importe_total = precio * cantidad 

        if producto not in ventas_resumen:
            ventas_resumen[producto] = {
                'total_precio': 0.0,
                'total_cantidad': 0,
                'total_importe': 0.0,
                'num_ventas': 0
            }

        ventas_resumen[producto]['total_precio'] += precio
        ventas_resumen[producto]['total_cantidad'] += cantidad
        ventas_resumen[producto]['total_importe'] += importe_total
        ventas_resumen[producto]['num_ventas'] += 1

    for producto, datos in ventas_resumen.items():
        datos['precio_promedio'] = datos['total_precio'] / datos['num_ventas']
    return ventas_resumen

def informe_final_ventas(resumen):
    productos_ordenados = sorted(resumen.keys())

    print(f"{'Producto':<30} {'Precio Promedio':<15} {'Unidades Vendidas':<18} {'Importe Total'}")
    print("=" * 75)

    for producto in productos_ordenados:
        datos = resumen[producto]
        print(f"{producto:<30} {datos['precio_promedio']:<15.2f} {datos['total_cantidad']:<18} {datos['total_importe']:.2f}")

datos = open("datos.dat", "r").readlines()
resumen = resumen_de_ventas(datos)
informe_final_ventas(resumen)

````
