## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
def verificarCumplimientoDeRangos(pCod, pNom, pPrec, pTip, pCant):
    pasoLaPrueba = True

    if(pCod != 4) and (len(pNom) < 1 or len(pNom) > 100)and (pPrec < 10 or pPrec > 10000) and (len(pTip) < 0 or len(pTip) > 100) and (pCant < 0 or pCant > 100):
        pasoLaPrueba = False

    return pasoLaPrueba


class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):   
            
        if verificarCumplimientoDeRangos(codigo, nombre, precio, tipo, cantidad):
            self._codigo = codigo
            self._nombre = nombre
            self._precio = precio
            self._tipo = tipo
            self._cantidad = cantidad
       

    @property
    def codigo(self):
        return self._codigo
    

    @codigo.setter
    def codigo(self, pCod):
        if pCod == 4:
            self._codigo = pCod


    @property
    def nombre(self):
        return self._nombre
    

    @nombre.setter
    def nombre(self, pNom):
        if len(pNom) > 0 and len(pNom) <= 100:
            self._nombre = pNom


    @property
    def precio(self):
        return self._precio
    
    
    @precio.setter
    def precio(self, pPrec):
        if pPrec >= 10 and pPrec <= 10000:
            self._precio = pPrec


    @property
    def cantidad(self):
        return self._cantidad
    

    @cantidad.setter
    def cantidad(self, pCant):
        if pCant >= 0 and pCant <= 100:
            self._cantidad = pCant


    @property
    def tipo(self):
        return self._tipo
    
    
    @tipo.setter
    def tipo(self, pTipo):
        self._tipo = pTipo

    
    def valorTotal(self):
        return self._precio * self._cantidad
    
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion='', codigos=[], tipos=[]):
        self.descripcion = descripcion
        self.codigos = codigos
        self.tipos = tipos

    def esAplicable(self, producto, cantidad):
        return (producto.codigo in self.codigos or producto.tipo in self.tipos) and cantidad >= 1

class OfertaDescuento(Oferta):
    def __init__(self, descuento, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.descuento = descuento

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return (producto.precio * self.descuento // 100) * cantidad
        return 0
    
    def aplicar(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            total = producto.precio * cantidad
            descuento = self.calcularDescuento(producto, cantidad)
            resultado = total - descuento
            return resultado
        return producto.precio * cantidad


class Oferta2x1(Oferta):
    def __init__(self, *args, **kwargs):

        super().__init__(*args, **kwargs)

    def calcularDescuento(self, producto, cantidad):

        if self.esAplicable(producto, cantidad):       
            resultadoCalculo = cantidad // 2  
            return resultadoCalculo * producto.precio
        return 0
    
    def aplicar(self, producto, cantidad):

        if self.esAplicable(producto, cantidad):    
            total = producto.precio * cantidad 
            descuento = self.calcularDescuento(producto, cantidad)
            resultado = total - descuento
            return resultado
        
        return producto.precio * cantidad  
  
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []

    @classmethod
    def leer(cls, archivo):
        catalogo = cls()
        
        with open(archivo, 'r') as file:
            datos = file.readlines()
            
            cabecera = datos[0].strip().split(',')
            if cabecera[0].lower() == 'codigo':
                datos = datos[1:]
            
            for dato in datos:

                if dato.strip():
                    codigo, nombre, precio, tipo, cantidad = dato.strip().split(',')
                    catalogo.agregar(Producto(codigo, nombre, float(precio), tipo, int(cantidad)))
        return catalogo

    def guardar(self, archivo):
        with open(archivo, 'w') as datos:
            for producto in self.productos:
                fila = f"{producto.codigo},{producto.nombre},{producto.precio},{producto.tipo},{producto.cantidad}\n"
                datos.write(fila)

    def agregar(self, *productos):
        for producto in productos:
            self.productos.append(producto)

    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None

    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    def buscarOferta(self, producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto, cantidad):
                return oferta
        return None

    def calcularDescuento(self, producto, cantidad):
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad)
        return 0

    def vender(self, producto, cantidad):
        if producto in self.productos and producto.cantidad >= cantidad:
            producto.cantidad -= cantidad
            return True
        return False

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        resultado = 0
        for producto in self.productos:
            resultado += producto.cantidad
        return resultado

    @property
    def valorTotal(self):
        resultado = 0
        for producto in self.productos:
            resultado += producto.precio * producto.cantidad
        return resultado

    def informe(self):
        tipos = set(producto.tipo for producto in self.productos)
        promedio = 0

        if self.cantidadUnidades > 0:
            promedio = self.valorTotal / self.cantidadUnidades
        else:
            promedio = 0

        detallesVenta = "Cantidad de productos: {cp} \n Cantidad de unidades: {cu} \n Precio Promedio: {p} \n Valor total: {vt} \n Tipos de productos: {t} \n Ofertas: \n Oferta 2x1:".format(cp = len(self.productos), cu = self.cantidadUnidades,p = promedio, vt = self.valorTotal,t = tipos)
        
        return detallesVenta
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del cat치logo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por c칩digo
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")

print(informe)
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
def esCorrectoElNombre(pNombre):
    resultado = False

    if(pNombre != "" and len(pNombre) > 0 and len(pNombre) <= 100):
        resultado = True
    return resultado

def esCorrectoElCuit(pCuit):
    resultado = True
    listCuit = list(pCuit)
    contador = 0

    if len(listCuit) == 13 and listCuit[2] == '-' and listCuit[11] == '-':
        while contador < len(listCuit):
            if contador != 2 and contador != 11:
                if not listCuit[contador].isnumeric():
                    resultado = False
            contador += 1
    else:
        resultado = False

    return resultado
             
class Cliente: 
    def __init__(self, nombre, cuit): 
        if esCorrectoElNombre(nombre) and esCorrectoElCuit(cuit):
            self._nombre = nombre
            self._cuit = cuit
        else:
            print("No es posible crear la instancia, datos incorrectos!")


    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, pNombre):
        if esCorrectoElNombre(pNombre):
            self._nombre = pNombre

    @property
    def cuit(self):
        return self._cuit

    @cuit.setter
    def cuit(self, pCuit):
        if esCorrectoElCuit(pCuit):
            self._cuit = pCuit
```

```python
class Factura:
    
    numero = 0

    def __init__(self, catalogo, cliente):
        self.numero = Factura.nuevoNumero()
        self.catalogo = catalogo
        self.cliente = cliente
        self.productos = {}
        

    #metodo para modificar un atributo a nivel de clase
    @classmethod
    def ultimaFactura(cls, pNumero):
        cls.numero = pNumero

    @classmethod
    def nuevoNumero(cls):
        cls.numero += 1
        return cls.numero

    def agregar(self, producto, cantidad):
        if cantidad <= 0:
            return 0
        
        productoCatalogo = self.catalogo.buscar(producto.codigo)
        if productoCatalogo:
            productoCatalogo.cantidad -= cantidad
        if producto.codigo in self.productos:
            self.productos[producto.codigo] += cantidad
        else:
            self.productos[producto.codigo] = cantidad

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        resultado = 0

        for value in self.productos.values():
            resultado += value

        return resultado

    @property
    def subtotal(self):
        resultado = 0

        for codigo, cantidad in self.productos.items():
            resultado += self.catalogo.buscar(codigo).precio * cantidad

        return resultado 

    @property
    def descuentos(self):
        resultado = 0

        for codigo, cantidad in self.productos.items():
            resultado += self.catalogo.calcularDescuento(self.catalogo.buscar(codigo), cantidad)

        return resultado

    @property
    def total(self):
        return self.subtotal - self.descuentos

    def imprimir(self):
        impresion = f"Numero de la factura: {self.numero}\n"
        impresion += f"Nombre del cliente: {self.cliente.nombre}\n"
        impresion += "Productos:\n"

        for codigo, cantidad in self.productos.items():
            producto = self.catalogo.buscar(codigo)
            impresion += f"{cantidad} {producto.nombre} - ${producto.precio}\n"
            Oferta = self.catalogo.buscarOferta(producto, cantidad)
            if Oferta:
                #verificamos si OfertaDescuento es de tipo Oferta
                if isinstance(Oferta, OfertaDescuento):
                    impresion += f"Descuento {Oferta.descuento}%\n"
                elif isinstance(Oferta, Oferta2x1):
                    impresion += "Oferta 2x1\n"
        impresion += "TOTAL:"
        impresion += f"{self.total:.2f}\n"
        return impresion
```

```python
## NO MODIFICAR ESTE CODIGO ##

# Prueba de la clase Factura #

# Creo un cat치logo con productos
catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)
catalogo.agregar(p1,p2,p3,p4)

# Registro ofertas
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

# Creo un cliente
cliente = Cliente('Juan Perez', '20-12345678-9')

# Creo una factura
Factura.ultimaFactura(100)
assert Factura.nuevoNumero() == 101
assert Factura.nuevoNumero() == 102

f1 = Factura(catalogo, cliente)
f1.agregar(p1, 5)
f1.agregar(p3, 3)

assert f1.numero == 103
assert f1.cantidadProductos == 2
assert f1.cantidadUnidades  == 8

# Agrega unidades de un producto ya agregado
f1.agregar(p1, 5)
assert f1.cantidadProductos == 2
assert f1.cantidadUnidades == 13

assert f1.subtotal   == 18600
assert f1.descuentos == 2700.0
assert f1.total == 15900.0

impresion = f1.imprimir()

assert "Juan Perez" in impresion
assert "10 Coca Cola" in impresion
assert "Sonrisa" in impresion
assert "Descuento 10%" in impresion
assert "Oferta 2x1" in impresion
assert "TOTAL:" in impresion
assert "15900.00" in impresion

print("Prueba pasada exitosamente!")
```
