## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo   =  tipo
        self.cantidad = cantidad

    @property
    def precio(self):
        return self.__precio
    
    @precio.setter
    def precio(self,valor):
        if valor < 0:
            return print(" Los valores de precio no pueden ser negativos :C ")
        else:
            self.__precio = valor

    @property
    def cantidad(self):
        return self.__cantidad
    
    @cantidad.setter
    def cantidad(self,valor):
        if valor < 0:
            return print("Las cantidades no pueden ser menores a 0 :c ")
        else:
            self.__cantidad = valor

    def valorTotal(self):
        return self.__cantidad * self.__precio
    
    def __str__(self) -> str:
        return "Producto {codigo}: {nombre} - ${precio} - {cantidad}".format(codigo=self.codigo, precio=self.precio, nombre=self.nombre, cantidad=self.cantidad)
   
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba de la clase Producto

p1 = Producto('0001', 'Coca Cola', 1500, 'gaseosa', 10)

assert p1.codigo == '0001'
assert p1.nombre == 'Coca Cola'
assert p1.precio == 1500

# Calcula el valor total 
assert p1.valorTotal() == 15000 

# Asegura que los valores de precio y cantidad sean validos
p1.precio = -200
assert p1.precio == 1500    # Rechaza el valor negativo

p1.precio = 2000 
assert p1.precio == 2000

p1.cantidad = -5
assert p1.cantidad == 10    # Rechaza el valor negativo

p1.cantidad = 50
assert p1.cantidad == 50
assert p1.valorTotal() == 100000

print("Prueba pasada exitosamente!")
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
import math

class Oferta:
    def __init__(self,descripcion, codigos = [], tipos = []):
        self.descripcion = descripcion
        self.cod_productos = codigos
        self.tipos_de_productos = tipos

    def esAplicable(self,producto,cantidad):
        codigos = getattr(self, 'cod_productos', [])
        tipos = getattr(self, 'tipos_de_productos', [])
        aplicar_descuento = producto.codigo in codigos or producto.tipo in tipos
        return aplicar_descuento
    
    def aplicar(self,producto,cantidad):
        return producto.precio * cantidad / 100
            

class OfertaDescuento(Oferta):
    def __init__(self, descuento, codigos = [], tipos = [], *args, **kwargs):
        self.descuento = descuento
        Oferta.__init__(self, f"Descuento {descuento}%", codigos, tipos)

    def calcularDescuento(self, producto, cantidadProducto):
        if self.esAplicable(producto, cantidadProducto):
            precioUnitarioOferta = self.aplicar(producto, self.descuento)
            return cantidadProducto * precioUnitarioOferta
        else:
            return 0
            
    
class Oferta2x1(Oferta):
    def __init__(self, codigos = [], tipos = [], *args, **kwargs):
        self.descuento = 50
        Oferta.__init__(self, "Oferta 2x1", codigos, tipos)

    def calcularDescuento(self, producto, cantidadProducto):
        if self.esAplicable(producto, cantidadProducto):
            precioUnitarioOferta = self.aplicar(producto, self.descuento)
            return math.floor(cantidadProducto / 2) * 2 * precioUnitarioOferta
        else:
            return 0
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv
import pprint
class Catalogo:
    def __init__(self) : 
        self.productos = []
        self.ofertas = []
        

    def leer(nombreArchivo):
        catalogo = Catalogo()
        with open(nombreArchivo, 'r') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader:
                catalogo.productos.append(Producto(row[0], row[1], int(row[2]), row[3], int(row[4])))
            return catalogo

    def guardar(self, nombreArchivo):
        with open(nombreArchivo, 'w') as file:
            writer = csv.writer(file)
            writer.writerow(["codigo","nombre","precio","tipo","cantidad"])
            for producto in self.productos:
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])

    def agregar(self, *productos):
        for producto in productos:
            self.productos.append(producto)


    def buscar(self,codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None

    def registrarOferta(self,oferta):
        self.ofertas.append(oferta)

    def buscarOferta(self,producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto, cantidad):
                return oferta
        return None
    
    def calcularDescuento(self, producto, cantidad): 
        oferta = self.buscarOferta(producto, cantidad)
        if oferta != None:
            return oferta.calcularDescuento(producto, cantidad)
        return 0
    
    def vender(self, producto, cantidadProducto):
        productoCatalogo = self.buscar(producto.codigo)
        if productoCatalogo.cantidad >= cantidadProducto:
            productoCatalogo.cantidad -= cantidadProducto

    @property
    def cantidadProductos(self):
        return len(self.productos)
    
    @property
    def cantidadUnidades(self):
        unidades = 0
        for producto in self.productos:
            unidades += producto.cantidad
        return unidades
    
    @property
    def valorTotal(self):
        total = 0
        for producto in self.productos:
            total += producto.cantidad * producto.precio
        return total
        
    @property
    def precioPromedio(self):
        if self.cantidadUnidades == 0:
            return 0
        return self.valorTotal / self.cantidadUnidades
        
    def informe(self):
        tipos = {}
        for producto in self.productos:
            if producto.tipo not in tipos:
                tipos[producto.tipo] = {'unidades':0,'valorTotal':0}
            tipos[producto.tipo]['unidades'] += producto.cantidad
            tipos[producto.tipo]['valorTotal'] += producto.cantidad * producto.precio
            
        informe = f"INFORME CATALOGO\n"
        informe += f"Cantidad de productos: {self.cantidadProductos}\n"
        informe += f"Cantidad de unidades: {self.cantidadUnidades}\n"
        informe += f"Precio Promedio: $ {self.precioPromedio:.2f}\n"
        informe += f"Valor total: $ {self.valorTotal:.2f}\n"
        informe += f"Tipos de productos: \n"
        for tipo, datos in tipos.items():
            precio_promedio_tipo = datos['valorTotal'] / datos['unidades'] if datos['unidades'] > 0 else 0
            informe += f"  - {tipo}: {datos['unidades']}u x $ {precio_promedio_tipo:.2f}\n"
        
        informe += f"Ofertas:\n"
        for oferta in self.ofertas:
            informe += f"  - {oferta.descripcion}\n"
        
        return informe
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del cat치logo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe

assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por c칩digo
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
import re

class Cliente: 
    def __init__(self, nombre, cuit): 
        self.__nombre = nombre
        self.__cuit = cuit
    @property
    def nombre(self):
        return self.__nombre
    @nombre.setter
    def nombre(self,nuevoNombre):
        if len(nuevoNombre) > 0 and len(nuevoNombre)<101:
            self.__nombre = nuevoNombre
    @property
    def cuit(self):
        return self.__cuit
    @cuit.setter
    def cuit(self,nuevoCuit):
        if len(nuevoCuit) == 15 and re.match(r'^\d{2}-\d{8}-\d{1}$', nuevoCuit):
            self.__cuit = nuevoCuit
    
```

```python
class Factura:
    def __init__(self, catalogo, cliente): 
        self.numero = Factura.nuevoNumero()
        self.catalogo = catalogo
        self.cliente = cliente
        self.items = []
        self.subtotal = 0
        self.descuentos = 0
        self.total = 0

    @classmethod
    def ultimaFactura(cls, numero):
        cls.contador = numero
    @classmethod
    def nuevoNumero(cls):
        cls.contador += 1
        return cls.contador
    def agregar(self,producto,cantidad):
        encontrado = False
        for item in self.items:
            if item['producto'].codigo == producto.codigo:
                encontrado = True
                itemActual = item
                cantidadActual = item['cantidad'] + cantidad
        oferta = self.catalogo.buscarOferta(producto,cantidad) 
        descuento = oferta.calcularDescuento(producto,cantidadActual if encontrado else cantidad) if oferta else 0
        if encontrado: 
            descuentoAnterior = oferta.calcularDescuento(producto,itemActual['cantidad']) if oferta else 0
            self.descuentos -= descuentoAnterior
            itemActual['cantidad'] = cantidadActual
            itemActual['subtotal'] = cantidadActual * producto.precio
            itemActual['descuento'] = descuento
        else: 
            self.items.append({
                "producto": producto,
                "cantidad": cantidad,
                "subtotal": producto.precio * cantidad,
                "descuento": descuento,
                "descripcionDescuento": oferta.descripcion if oferta else ""
            })
        self.subtotal += producto.precio * cantidad
        self.descuentos += descuento
        self.total = self.subtotal - self.descuentos
        productoCatalogo = self.catalogo.buscar(producto.codigo)
        productoCatalogo.cantidad -= cantidad

    @property
    def cantidadProductos(self):
        return len(self.items)
    
    @property
    def cantidadUnidades(self):
        unidades = 0
        for item in self.items:
            unidades += item['cantidad']
        return unidades

    def imprimir(self):
        factura = f"Factura: {self.numero}\n"
        factura += f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})\n\n"
        
        for item in self.items:
            producto = item["producto"]
            factura += f"- {item['cantidad']} {producto.nombre:<30} x ${producto.precio:<6} = ${item['subtotal']:<6}\n"
            if item["descuento"] > 0:
                factura += f"      {item['descripcionDescuento']}:                        - ${item['descuento']:<6}\n"
        
        factura += "\n"
        factura += f"{'Subtotal:':>50} ${self.subtotal:<6}\n"
        factura += f"{'Descuentos:':>50} ${self.descuentos:<6}\n"
        factura += f"{'-----------------------':>50}\n"
        factura += f"{'TOTAL:':>50} ${self.total:.2f}\n"
        
        return factura
    
```
